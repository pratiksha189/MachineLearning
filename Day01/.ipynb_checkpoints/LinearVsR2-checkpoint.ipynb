```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import  train_test_split
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error,root_mean_squared_error
import os
from sklearn.preprocessing import PolynomialFeatures
```

```python
pizza = pd.read_csv('pizza.csv')
pizza
```

```python
X = pizza[['Promote']]
y = pizza['Sales']
lr = LinearRegression()
lr.fit(X,y)
lr.coef_, lr.intercept_
```

```python
Boston = pd.read_csv('Boston.csv')
Boston
```

```python

X=Boston.drop("medv",axis=1)
y=Boston['medv']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=26)
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
print("MAE=",mean_absolute_error(y_test,y_pred))
print("RMSE=",root_mean_squared_error(y_test,y_pred))
print("R2=",r2_score(y_test,y_pred))
```

```python
# for degree 2
poly=PolynomialFeatures(degree=2,include_bias=False).set_output(transform="pandas")
X_poly=poly.fit_transform(X)
lr=LinearRegression()
lr.fit(X_poly,y)
lr.coef_,lr.intercept_
```

```python
# for degree 3
poly=PolynomialFeatures(degree=3,include_bias=False).set_output(transform="pandas")
X_poly=poly.fit_transform(X)
lr=LinearRegression()
lr.fit(X_poly,y)
lr.coef_,lr.intercept_
```

# BOSTON

```python
Boston = pd.read_csv('Boston.csv')
Boston
```

```python
# for degree 2
X=Boston.drop("medv",axis=1)
y=Boston['medv']
poly=PolynomialFeatures(degree=2,include_bias=False).set_output(transform="pandas")
X_poly=poly.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_poly,y,test_size=0.3,random_state=26)   # change is only (X_poly)
# for degree 3
lr=LinearRegression()

lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
print("R2=",r2_score(y_test,y_pred))
```

```python
# for degree 3
Boston = pd.read_csv('Boston.csv')
Boston

X=Boston.drop("medv",axis=1)
y=Boston['medv']
poly=PolynomialFeatures(degree=3,include_bias=False).set_output(transform="pandas")
X_poly=poly.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_poly,y,test_size=0.3,random_state=26)   # change is only (X_poly)
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
print("R2=",r2_score(y_test,y_pred))
```

# housing dataset

```python
housing = pd.read_csv('Housing.csv')
housing
```

```python
# for degree 1
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=26)
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
# print("MAE=",mean_absolute_error(y_test,y_pred))
# print("RMSE=",root_mean_squared_error(y_test,y_pred))
print("R2=",r2_score(y_test,y_pred))
```

```python
# for degree 2
dum_hous=pd.get_dummies(housing,dtype=int,drop_first=True)
dum_hous
X,y=dum_hous.drop("price",axis=1),dum_hous["price"]
poly=PolynomialFeatures(degree=2,include_bias=False).set_output(transform="pandas")
X_poly=poly.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_poly,y,test_size=0.3,random_state=26)   # change is only (X_poly)
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
print("R2=",r2_score(y_test,y_pred))
```

```python
# for degree 3
dum_hous=pd.get_dummies(housing,dtype=int,drop_first=True)
dum_hous
X,y=dum_hous.drop("price",axis=1),dum_hous["price"]
poly=PolynomialFeatures(degree=3,include_bias=False).set_output(transform="pandas")
X_poly=poly.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_poly,y,test_size=0.3,random_state=26)   # change is only (X_poly)
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
print("R2=",r2_score(y_test,y_pred))
```
