```python
housing = pd.read_csv("Housing.csv")
housing
```

## column transformer

```python

ohe = OneHotEncoder(sparse_output=False, drop="first").set_output(transform="pandas")
scalar=StandardScaler()
trans = ColumnTransformer(
    transformers=[("OHE", ohe, ["driveway", "recroom", "fullbase", "gashw", "airco", "prefarea"]),("STD",scalar,["price","lotsize"])],remainder="passthrough",
    verbose_feature_names_out=False).set_output(transform="pandas")
trans.fit_transform(housing)
```

#### make_column_selector

```python
from sklearn.compose import make_column_selector

trans = ColumnTransformer(
    transformers=[("OHE", ohe,make_column_selector(dtype_include=object))],remainder="passthrough",
    verbose_feature_names_out=False).set_output(transform="pandas")
trans.fit_transform(housing)
```
