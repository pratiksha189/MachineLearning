```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import  ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, root_mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression, Ridge
import os
os.chdir("D:/Machine_Learning/Datasets")
```

```python
housing = pd.read_csv("Housing.csv")
housing
```

```python
X = housing.drop('price',axis=1)
y=housing['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=26)
```

```python
from sklearn.compose import make_column_selector
ohe=OneHotEncoder(sparse_output=False,drop="first").set_output(transform="pandas")
ohe.fit_transform(housing)
scalar=StandardScaler()
trans = ColumnTransformer(
    transformers=[("OHE", ohe,make_column_selector(dtype_include=object))],remainder="passthrough",
    verbose_feature_names_out=False).set_output(transform="pandas")
trans.fit_transform(housing)
```

```python
X_trn_trans=trans.fit_transform(X_train)
X_tst_trans=trans.transform(X_test)
```

```python
lr=LinearRegression()
lr.fit(X_trn_trans,y_train)
y_pred =lr.predict(X_tst_trans)
print("R2 =",r2_score(y_test,y_pred))
```

```python
ridge = Ridge()
ridge.fit(X_trn_trans, y_train)
y_pred = ridge.predict(X_tst_trans)
print("R2 Score =",r2_score(y_test, y_pred))
```
