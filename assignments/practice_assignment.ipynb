```python
train = pd.read_csv('train.csv')
train
```

```python
X = train.drop("diagnosed_diabetes", axis=1)
y = train['diagnosed_diabetes']
y.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=26)
```

```python
from sklearn.compose import make_column_selector

# 1. Set sparse_output=True (default) and remove .set_output(transform="pandas")
ohe = OneHotEncoder(drop="first", sparse_output=True)

# 2. Build the transformer without forcing pandas output
trans = ColumnTransformer(
    transformers=[
        ("OHE", ohe, make_column_selector(dtype_include=object))
    ], 
    remainder="passthrough",
    verbose_feature_names_out=False
)

# 3. Fit and transform (this will now be very fast and memory-efficient)
X_trn_sparse = trans.fit_transform(X_train)
X_tst_sparse = trans.transform(X_test)
```

```python
X_trn_ohe = trans.fit_transform(X_train)
X_tst_ohe = trans.transform(X_test)
```

```python
lr = LogisticRegression() 
lr.fit(X_trn_ohe, y_train)
y_pred_prob = lr.predict_proba(X_tst_ohe)
log_loss(y_test,y_pred_prob)
```

```python
solvers = ["lbfgs", "newton-cg", "newton-cholesky", "sag", "saga"]
Cs = np.linspace(0.001, 5, 20)
scores = []

# 1. Added the missing solver loop
for s in solvers:
    for c in Cs:
        # 2. Use the variable 's' from the loop
        lr = LogisticRegression(solver=s, C=c, max_iter=1000)
        lr.fit(X_trn_ohe, y_train) # Ensure you use y_train, not the full y
        
        # 3. Use predict for F1 score (F1 needs classes, not probabilities)
        y_pred = lr.predict(X_tst_ohe)
        f1 = f1_score(y_test, y_pred, average='macro')
        
        # 4. Fixed the syntax error: changed append to take a list [s, c, f1]
        scores.append([s, c, f1])

df_scores = pd.DataFrame(scores, columns=['solver', 'C', 'score'])
print(df_scores.sort_values('score', ascending=False).head())
```
