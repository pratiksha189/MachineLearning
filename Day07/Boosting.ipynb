```python
cancer=pd.read_csv(r"../Datasets/BreastCancer.csv")
cancer
```

```python
X = cancer.drop(['Code','Class'],axis=1)
y=cancer['Class']
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.3,random_state=26,stratify = cancer['Class'])
```

```python
dtc=DecisionTreeClassifier(max_depth = 1,random_state=26)
knn = KNeighborsClassifier(n_neighbors=3)
lr = LogisticRegression()
lda = LinearDiscriminantAnalysis()
nb = GaussianNB() 
ada=AdaBoostClassifier(random_state=26,estimator=dtc)
ada.fit(X_train,y_train)
y_pred_prob =ada.predict_proba(X_test)
y_pred=ada.predict(X_test)
                                                                                                                                                                                                                               
print("accuracy:",accuracy_score(y_test,y_pred))
print("log loss:",log_loss(y_test,y_pred_prob))
```

```python
hr = pd.read_csv(r"../Datasets/HR_comma_sep.csv")
hr
```

```python
X = hr.drop('left',axis=1)
y=hr['left']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=26, stratify=hr['left'])
```

```python
ohe=OneHotEncoder(sparse_output=False,drop="first").set_output(transform="pandas")
trans = ColumnTransformer(
    transformers=[("OHE", ohe,make_column_selector(dtype_include=object))],remainder="passthrough",
    verbose_feature_names_out=False).set_output(transform="pandas")
X_trn_ohe = trans.fit_transform(X_train)
X_tst_ohe = trans.transform(X_test)
```

```python
dtc=DecisionTreeClassifier(max_depth = 1,random_state=26)
lr = LogisticRegression()
nb = GaussianNB() 
for e in [dtc,lr,nb]:
    ada=AdaBoostClassifier(random_state=26,estimator=dtc)
    ada.fit(X_trn_ohe,y_train)
    y_pred_prob =ada.predict_proba(X_tst_ohe)
    y_pred=ada.predict(X_tst_ohe)
    print("for:",e,":")                                                                                                                                                                                                                              
    print("accuracy:",accuracy_score(y_test,y_pred))
    print("log loss:",log_loss(y_test,y_pred_prob))
```

```python


connc_data = pd.read_csv(r'../Datasets/Concrete_data.csv')
connc_data
```

```python
X,y=connc_data.drop("Strength",axis=1),connc_data["Strength"]
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=26)
```

```python
dtc=DecisionTreeRegressor(max_depth = 1,random_state=26)
lr = LinearRegression()
en = ElasticNet(random_state=26) 
la = Lasso(random_state=26)
ri = Ridge(random_state=26)

for e in [dtc,lr,en,la,ri]:
    ada=AdaBoostRegressor(random_state=26,n_estimators=100,estimator=e)
    ada.fit(X_train,y_train)
    y_pred=ada.predict(X_test)

    print("for:",e,":")                                                                                                                                                                                                                              
    print("r2:",r2_score(y_test,y_pred))
   
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error

n_est = [10, 25, 50, 75, 100]
rate = np.linspace(0.0001, 0.9, 30)
depth = [2, 3, 4, 5, 6]
scores = []

for n in n_est:
    for r in rate:
        for d in depth:
            gbm = GradientBoostingRegressor(random_state=26, n_estimators=n, learning_rate=r, max_depth=d)
            gbm.fit(X_train, y_train)
            y_pred = gbm.predict(X_test)
            mse = mean_squared_error(y_test, y_pred)
            scores.append([n, r, d, mse])

df_scores = pd.DataFrame(scores, columns=['n_est', 'rate', 'depth', 'score'])
df_scores = df_scores.sort_values('score')
print(df_scores.head())
```
