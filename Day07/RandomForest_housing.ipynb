```python
housing = pd.read_csv(r"../Datasets/Housing.csv")
housing
```

```python
dum_hous = pd.get_dummies(housing, drop_first=True)
X = dum_hous.drop('price', axis=1)
y = dum_hous['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=26)
```

```python
from sklearn.compose import make_column_selector,ColumnTransformer

ohe=OneHotEncoder(sparse_output=False,drop="first").set_output(transform="pandas")
trans = ColumnTransformer(
    transformers=[("OHE", ohe,make_column_selector(dtype_include=object))],remainder="passthrough",
    verbose_feature_names_out=False).set_output(transform="pandas")
```

```python
X_trn_ohe = trans.fit_transform(X_train)
X_tst_ohe = trans.transform(X_test)
```

```python
features = [2,3,4,5,6,7]
scores = []
for f in tqdm(features) :
    for n in [25,50,75,100,125]:
         rf = RandomForestClassifier(random_state=26,max_features=f,n_estimators=n)
         rf.fit(X_trn_ohe,y_train)
         y_pred_prob = rf.predict_proba(X_tst_ohe)
         scores.append([f,n,log_loss(y_test,y_pred_prob)])
df_score = pd.DataFrame(scores,columns = ['max_features','n_estimators','log_loss'])
df_score.sort_values('log_loss',ascending = True)
```
