```python
# Perform Linear Regression by filling missing values using mean and median
```

```python
# Sir's way 
X = chem.drop('Yield',axis=1)
y=chem['Yield']

mean_imp = SimpleImputer(strategy='mean')
X_imp = mean_imp.fit_transform(X)

X_train,X_test,y_train,y_test = train_test_split(X_imp, y, test_size=0.3, random_state=26)
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred =lr.predict(X_test)

print("R2 =",r2_score(y_test,y_pred))
```

```python
mean_imp = SimpleImputer(strategy='mean').set_output(transform="pandas")
mean_imp.fit(chem) # Calculates mean of every column
chem1 = mean_imp.transform(chem)
chem1
```

```python
chem1.isnull().sum().sum()
```

```python
X = chem1.drop('Yield',axis=1)
y=chem1['Yield']

X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.3, random_state=26)
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred =lr.predict(X_test)

print("R2 =",r2_score(y_test,y_pred))
```

```python
med_imp = SimpleImputer(strategy='median').set_output(transform="pandas")
med_imp.fit(chem) # Calculates mean of every column
chem2 = med_imp.transform(chem)
chem2
```

```python
X = chem2.drop('Yield',axis=1)
y=chem2['Yield']

X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.3, random_state=26)
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred =lr.predict(X_test)

print("R2 =",r2_score(y_test,y_pred))
```

### The following is the way to write above code correctly

```python
X = chem.drop('Yield',axis=1)
y=chem['Yield']
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.3, random_state=26)

mean_imp = SimpleImputer(strategy='mean').set_output(transform="pandas")
X_trn_imp = mean_imp.fit_transform(X_train)
X_tst_imp = mean_imp.transform(X_test)

lr.fit(X_trn_imp,y_train)
y_pred =lr.predict(X_tst_imp)

print("R2 =",r2_score(y_test,y_pred))
```

```python
med_imp = SimpleImputer(strategy='median').set_output(transform="pandas")

X_trn_imp = med_imp.fit_transform(X_train)
X_tst_imp = med_imp.transform(X_test)

lr.fit(X_trn_imp, y_train)
y_pred =lr.predict(X_tst_imp)

print("R2 =",r2_score(y_test,y_pred))
```
