```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, root_mean_squared_error, r2_score

import os
os.chdir("D:/Machine_Learning/Datasets") #fetch the dataset from the folder
os.chdir("D:/Machine_Learning/Cases/Concrete_Strength")

from sklearn.preprocessing import PolynomialFeatures
```

### Boston data

```python
boston=pd.read_csv("Boston.csv")
boston  # medv-dependent variable
```

```python
X = boston.drop('medv',axis=1)
y=boston['medv']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=26)
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred =lr.predict(X_test)
print("MAE =",mean_absolute_error(y_test,y_pred))
print("RMSE =",root_mean_squared_error(y_test,y_pred))
print("R2 =",r2_score(y_test,y_pred))
```

### Polynomial  class Pizza data

```python
X = pizza[['Promote']]  # Dataframe
y = pizza['Sales']      # Series
```

```python
poly = PolynomialFeatures(degree =2, include_bias=False).set_output(transform="pandas") # include_bias=False removes the constant values
X_poly = poly.fit_transform(X)
X_poly.head()
```

```python
lr = LinearRegression()
lr.fit(X_poly, y)
lr.coef_ , lr.intercept_
# array([ 2.48006959e+01, -1.30947333e-02]) -> x(promote)= 2.48006959e+01  x^2(promote^2)=-1.30947333e-02]

# sales = 24.8promote - -0.0130promote^2 - 18.274  (Polynomial Equation)
```

## Housing Dataset

```python
housing = pd.read_csv('Housing.csv')
housing
```

```python
dum_hous = pd.get_dummies(housing, dtype=int, drop_first=True)  # drop_first=True drops the one col as two cols created for yes and no
dum_hous
```

```python
X, y = dum_hous.drop('price', axis=1), dum_hous['price']
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.3, random_state=26)

# Simple Linear regression (degree 1)
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred =lr.predict(X_test)
# print("MAE =",mean_absolute_error(y_test,y_pred))
# print("RMSE =",root_mean_squared_error(y_test,y_pred))
print("R2 =",r2_score(y_test,y_pred))
```

```python
# for degree 2
poly = PolynomialFeatures(degree =2, include_bias=False).set_output(transform="pandas") 
X_poly = poly.fit_transform(X)
X_poly

X_train,X_test,y_train,y_test = train_test_split(X_poly, y, test_size=0.3, random_state=26)

lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred =lr.predict(X_test)
# print("MAE =",mean_absolute_error(y_test,y_pred))
# print("RMSE =",root_mean_squared_error(y_test,y_pred))
print("R2 =",r2_score(y_test,y_pred))
```

```python
# for degree 3
poly = PolynomialFeatures(degree =3, include_bias=False).set_output(transform="pandas") 
X_poly = poly.fit_transform(X)
X_poly

X_train,X_test,y_train,y_test = train_test_split(X_poly, y, test_size=0.3, random_state=26)

lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred =lr.predict(X_test)
# print("MAE =",mean_absolute_error(y_test,y_pred))
# print("RMSE =",root_mean_squared_error(y_test,y_pred))
print("R2 =",r2_score(y_test,y_pred))
```
