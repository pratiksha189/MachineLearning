```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
x = np.linspace(-10,10)  #array
y = 3*x+5   #array
x
```

### Method of Least Squares

```python
sum_xy = np.sum(x*y)
n=len(x)
xbar,ybar=np.mean(x), np.mean(y)
sum_x2=np.sum(x*x)
b1=((sum_xy/n) - (xbar*ybar)) / ((sum_x2/n) - (xbar*xbar))
b0=ybar - b1*xbar
b0,b1
```

```python
lr=LinearRegression()
lr.fit(x.reshape(-1,1), y)  #reshape becz x(50,) not considerable in scikit it should be X(50,1) matrix with 50 rows and 1 col
lr.coef_, lr.intercept_
```
