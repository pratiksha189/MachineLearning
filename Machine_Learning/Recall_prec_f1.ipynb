```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_absolute_error, root_mean_squared_error, r2_score
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score
from sklearn.metrics import recall_score, precision_score, classification_report

from sklearn.impute import SimpleImputer 

import os
os.chdir("D:/Machine_Learning/Datasets") #fetch the dataset from the folder
```

```python
sonar = pd.read_csv('D:/Machine_Learning/Cases/Sonar/Sonar.csv')
sonar
```

```python
X , y = sonar.drop('Class', axis=1), sonar['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state=26, stratify= sonar['Class'])

lr= LogisticRegression()
lr.fit(X_train, y_train)
y_pred_prob = lr.predict_proba(X_test)
y_pred_prob   # probability p[y==0] , p[y==1]
                # give output as 0,1 based on max prob...
```

```python
y_pred = lr.predict(X_test)
pd.crosstab(y_test, y_pred, margins=True)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(recall_score(y_test, y_pred, pos_label='M'))  # pos_label==Positive label
print(recall_score(y_test, y_pred, pos_label='R'))
```

```python
print(precision_score(y_test, y_pred, pos_label='M'))
print(precision_score(y_test, y_pred, pos_label='R'))
```

```python
print(f1_score(y_test, y_pred, pos_label='M'))
print(f1_score(y_test, y_pred, pos_label='R'))
```

```python
# All in one
print(classification_report(y_test, y_pred))  # without positive label

# Macro avg = 
# weighted avg = (0.88*34 + 0.81*29)/63
```
